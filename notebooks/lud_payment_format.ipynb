## Removing Reinvestments and Wire transactions

```python
import pandas as pd 
```

```python
 df= pd.read_csv('/Users/ludwingmoncadabellorin/code/Malvinrud/aml_detector/raw_data/new_hi_small_trans.csv')
```

```python
df.head()
```

```python
# Identify Unique Payment Format 
unique_currencies = df['Payment Format'].unique()
print(unique_currencies)
```

```python
# Get the total count of each payment format
payment_format_counts = df['Payment Format'].value_counts()

# Calculate the percentages
payment_format_percentages = payment_format_counts / len(df) * 100

# Get the counts of laundering and non-laundering transactions for each payment format
laundering_counts = df[df['Is Laundering'] == 1]['Payment Format'].value_counts()
non_laundering_counts = df[df['Is Laundering'] == 0]['Payment Format'].value_counts()

# Calculate the percentages
laundering_percentages = laundering_counts / payment_format_counts * 100
non_laundering_percentages = non_laundering_counts / payment_format_counts * 100

# Display the counts and percentages side by side
df_counts_percentages = pd.DataFrame({
    'Count': payment_format_counts,
    'Percentage of Total': payment_format_percentages,
    'Laundering Count': laundering_counts,
    'Laundering Percentage': laundering_percentages,
    'Non-Laundering Count': non_laundering_counts,
    'Non-Laundering Percentage': non_laundering_percentages
})
df_counts_percentages
```

1) Reinvestment represents only 9% of the dataset and non of them is laundering
2) Wire only represents 3% of the dataset and non of them is laundering.
**Proceeding to delete them**

```python
# List of payment formats to remove
remove_formats = ['Reinvestment', 'Wire']

# Remove the rows with the specified payment formats
df = df[~df['Payment Format'].isin(remove_formats)]
```

```python
#Re run everything for good measure ;)

# Get the total count of each payment format
payment_format_counts = df['Payment Format'].value_counts()

# Calculate the percentages
payment_format_percentages = payment_format_counts / len(df) * 100

# Get the counts of laundering and non-laundering transactions for each payment format
laundering_counts = df[df['Is Laundering'] == 1]['Payment Format'].value_counts()
non_laundering_counts = df[df['Is Laundering'] == 0]['Payment Format'].value_counts()

# Calculate the percentages
laundering_percentages = laundering_counts / payment_format_counts * 100
non_laundering_percentages = non_laundering_counts / payment_format_counts * 100

# Display the counts and percentages side by side
df_counts_percentages = pd.DataFrame({
    'Count': payment_format_counts,
    'Percentage of Total': payment_format_percentages,
    'Laundering Count': laundering_counts,
    'Laundering Percentage': laundering_percentages,
    'Non-Laundering Count': non_laundering_counts,
    'Non-Laundering Percentage': non_laundering_percentages
})
df_counts_percentages
```
